```python
import pandas as pd
```

```python
race_data = pd.read_csv("Resources/MERGED2016_17_PP.csv")
race_data.head()
```

```python
#  
flag_df = race_data[['UNITID','INSTNM','HBCU', 'PBI', 'ANNHI','TRIBAL','AANAPII','HSI', 'NANTI']]
new_flag_df = flag_df.rename(index=str, columns={"PBI":"Pred_Black_Inst", 
                                                 "ANNHI": "Alaska/Hawaiian", 
                                                 "TRIBAL":"Tribal",
                                                 "AANAPII":"Asian/Native/Pacific_Islander", 
                                                 "HSI":"Hispanic_Serving", 
                                                 "NANTI":"Native_American_Non_Tribal"})

# Undergraduate degree-seeking 
ugds_df = race_data[['UNITID','INSTNM','UGDS_WHITE', 'UGDS_BLACK', 'UGDS_HISP', 'UGDS_ASIAN', 'UGDS_AIAN', 'UGDS_NHPI', 'UGDS_2MOR', 'UGDS_NRA','UGDS_UNKN','UGDS_WHITENH', 'UGDS_BLACKNH', 'UGDS_API']]
new_ugds_df = ugds_df.rename(index=str, columns={'UGDS_WHITE':'White', 
                                                 'UGDS_BLACK':'Black', 
                                                 'UGDS_HISP':'Hispanic', 
                                                 'UGDS_ASIAN':'Asian', 
                                                 'UGDS_AIAN':'American_Indian/Alaska_Native', 
                                                 'UGDS_NHPI':'Native_Hawaiian/Pacific_Islander', 
                                                 'UGDS_2MOR':'Two_or_More_Races', 
                                                 'UGDS_NRA': 'Non_Resident_Aliens',
                                                 'UGDS_UNKN': 'Unknown',
                                                 'UGDS_WHITENH': 'White_Non_Hispanic', 
                                                 'UGDS_BLACKNH': 'Black_Non_Hispanic', 
                                                 'UGDS_API':'Asian/Pacific_Islander'})

# Completion rate for first-time, full-time students at four-year institutions (150% of expected time to completion)
completion_df = race_data[['UNITID','INSTNM', 'C150_4_WHITE', 'C150_4_BLACK', 'C150_4_HISP', 'C150_4_ASIAN', 'C150_4_AIAN', 'C150_4_NHPI', 'C150_4_2MOR', 'C150_4_NRA','C150_4_UNKN','C150_4_WHITENH', 'C150_4_BLACKNH', 'C150_4_API']]
new_completion_df = completion_df.rename(columns={'C150_4_WHITE':'White', 
                                                  'C150_4_BLACK':'Black', 
                                                  'C150_4_HISP': 'Hispanic', 
                                                  'C150_4_ASIAN':'Asian', 
                                                  'C150_4_AIAN':'American_Indian/Alaska_Native', 
                                                  'C150_4_NHPI':'Native_Hawaiian/Pacific_Islander', 
                                                  'C150_4_2MOR':'Two_or_More_Races', 
                                                  'C150_4_NRA':'Non_Resident_Aliens',
                                                  'C150_4_UNKN':'Unknown',
                                                  'C150_4_WHITENH':'White_Non_Hispanic', 
                                                  'C150_4_BLACKNH':'Black_Non_Hispanic', 
                                                  'C150_4_API':'Asian/Pacific_Islander'}])

# Adjusted cohort count for completion rate at four-year institutions (denominator of 150% completion rate)
cohort_df = race_data[['UNITID','INSTNM', 'D150_4_WHITE', 'D150_4_BLACK', 'D150_4_HISP', 'D150_4_ASIAN', 'D150_4_AIAN', 'D150_4_NHPI', 'D150_4_2MOR', 'D150_4_NRA', 'D150_4_UNKN']]
new_cohort_df = cohort_df(columns={'D150_4_WHITE':'White', 
                                   'D150_4_BLACK':'Black', 
                                   'D150_4_HISP':'Hispanic', 
                                   'D150_4_ASIAN':'Asian', 
                                   'D150_4_AIAN':'American_Indian/Alaska_Native', 
                                   'D150_4_NHPI':'Native_Hawaiian/Pacific_Islander', 
                                   'D150_4_2MOR':'Two_or_More_Races', 
                                   'D150_4_NRA':'Non_Resident_Aliens', 
                                   'D150_4_UNKN':'Unknown'})

new_flag_df.to_csv('Output/race_flag.csv')
new_ugds_df.to_csv('Output/race_ugds.csv')
new_completion_df.to_csv('Output/race_completion.csv')
new_cohort_df.to_csv('Output/race_cohort.csv')
```
