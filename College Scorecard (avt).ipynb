```python
#Dependencies 
import pandas as pd
```

```python
#Load the CSV file
load_file = "Resources/v2_MERGED2016_17_PP.csv"
```

```python
#Create DataFrame from CSV file
college_df = pd.read_csv(load_file)
```

```python
#List all the columns 
college_df.columns
```

```python
#Print the original DataFrame
college_df.head()
```

```python
#Create new DataFrame with debt and repayment information.
debt_repymt_df = college_df[['UNITID', 'Institution_Name', 'Degree_Granted', 'Female_7_Year_Repayment_Rate', 'Male_7_Year_Repayment_Rate', '1St_Gen_7_Year_Repayment_Rate', 'Non_1St_Gen_7_Year_Repayment_Rate', 
                             'Median_Debt_For_Female_Students', 'Median_Debt_For_Male_Students', 'Median_Debt_For_1St_Gen_Students', 'Median_Debt_For_Non_1St_Gen_Students']]
debt_repymt_df.head()
```

```python
#Revise wording of values
debt_repymt_df.loc[debt_repymt_df.Degree_Granted == 0, 'Predom_DegreeGranting'] = 'Not Classified'  
debt_repymt_df.loc[debt_repymt_df.Degree_Granted == 1, 'Predom_DegreeGranting'] = 'Predominantly Cert-Degree Granting'  
debt_repymt_df.loc[debt_repymt_df.Degree_Granted == 2, 'Predom_DegreeGranting'] = 'Predominantly Assoc-Degree Granting' 
debt_repymt_df.loc[debt_repymt_df.Degree_Granted == 3, 'Predom_DegreeGranting'] = 'Predominantly BS-Degree Granting' 
debt_repymt_df.loc[debt_repymt_df.Degree_Granted == 4, 'Predom_DegreeGranting'] = 'Entirely Grad-Degree Granting' 
debt_repymt_df.head()
```

```python
#Remove Degree_Granted column
del debt_repymt_df['Degree_Granted']
debt_repymt_df.head()
```

```python
#Reorganize columns
debt_repymt2_df = debt_repymt_df[['UNITID', 'Institution_Name', 'Predom_DegreeGranting', 'Median_Debt_For_Female_Students', 'Median_Debt_For_Male_Students', 'Median_Debt_For_1St_Gen_Students', 
                                 'Median_Debt_For_Non_1St_Gen_Students', 'Female_7_Year_Repayment_Rate', 'Male_7_Year_Repayment_Rate', '1St_Gen_7_Year_Repayment_Rate', 'Non_1St_Gen_7_Year_Repayment_Rate']]
debt_repymt2_df.head()
```

```python
#Rename columns
renamed_debtRepymt_df = debt_repymt2_df.rename(columns={'Median_Debt_For_Female_Students': 'Med Student Debt (Female)', 'Median_Debt_For_Male_Students': 'Med Student Debt (Male)',
                                                       'Median_Debt_For_1St_Gen_Students': 'Med Student Debt (First Gen)', 'Median_Debt_For_Non_1St_Gen_Students': 'Med Student Debt (Non-First Gen)', 'Female_7_Year_Repayment_Rate': '7yr Repymt Rate (Female)', 
                                                        'Male_7_Year_Repayment_Rate': '7yr Repymt Rate (Male)', '1St_Gen_7_Year_Repayment_Rate': '7yr Repymt Rate (First Gen)', 'Non_1St_Gen_7_Year_Repayment_Rate': '7yr Repymt Rate (Non-Firs Gen)'})
renamed_debtRepymt_df.head()
```

```python
#Create new DataFrame with degree and major information.
degMaj_df = college_df[['UNITID', 'Institution_Name', 'Degree_Granted', 'Communication_Journalism', 'Computer_And_Information_Sciences', 'Engineering', 'Liberal_Arts', 'Math_And_Statistics',
                            'Social_Sciences', 'Business']]
degMaj_df.head()
```

```python
#Revise wording of values
degMaj_df.loc[degMaj_df.Degree_Granted == 0, 'Predom_DegreeGranting'] = 'Not Classified'  
degMaj_df.loc[degMaj_df.Degree_Granted == 1, 'Predom_DegreeGranting'] = 'Predominantly Cert-Degree Granting'  
degMaj_df.loc[degMaj_df.Degree_Granted == 2, 'Predom_DegreeGranting'] = 'Predominantly Assoc-Degree Granting' 
degMaj_df.loc[degMaj_df.Degree_Granted == 3, 'Predom_DegreeGranting'] = 'Predominantly BS-Degree Granting' 
degMaj_df.loc[degMaj_df.Degree_Granted == 4, 'Predom_DegreeGranting'] = 'Entirely Grad-Degree Granting' 
degMaj_df.head()
```

```python
#Remove Degree_Granted column
del degMaj_df['Degree_Granted']
degMaj_df.head()
```

```python
#Reorganize columns
degMaj2_df = degMaj_df[['UNITID', 'Institution_Name', 'Predom_DegreeGranting', 'Communication_Journalism', 'Computer_And_Information_Sciences', 'Engineering', 'Liberal_Arts', 
                                 'Math_And_Statistics', 'Social_Sciences', 'Business']]
degMaj2_df.head()
```

```python
#Rename columns
renamed_degMajor_df = degMaj2_df.rename(columns={'Communication_Journalism':'% Degree Granted (Comm & Journalism)', 'Computer_And_Information_Sciences':'% Degree Granted (Comp & Info Sci)',
                                                       'Engineering': '% Degree Granted (Engineering)', 'Liberal_Arts': '% Degree Granted (Lib Arts)', 'Math_And_Statistics': '% Degree Granted (Math & Stats)', 
                                                        'Social_Sciences': '% Degree Granted (Soc Sci)', 'Business': '% Degree Granted (Bus)'})
renamed_degMajor_df.head()
```

```python
# Export file as a CSV, without the Pandas index, but with the header
renamed_debtRepymt_df.to_csv("Output/debtRepymt.csv", index=False, header=True)
```

```python
# Export file as a CSV, without the Pandas index, but with the header
renamed_degMajor_df.to_csv("Output/degreeMajor.csv", index=False, header=True)
```
